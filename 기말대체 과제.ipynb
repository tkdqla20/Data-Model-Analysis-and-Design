```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("A회사_보습제_매출데이터_v1.csv", encoding="cp949")
df
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "C:\\Windows\\Fonts\\malgun.ttf"  # 한글 폰트 파일 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 'Unnamed: 0' 컬럼을 인덱스로 설정
df.set_index('Unnamed: 0', inplace=True)

# 문자열을 숫자로 변환
df = df.applymap(lambda x: float(x.replace(',', '').replace('$', '')))
```

```python
# 월별 총매출 추출
monthly_sales = df.loc["총매출"].to_list()

# 그래프 그리기
plt.plot(monthly_sales, label="월별 총매출")
plt.title("월별 총매출 그래프")
plt.xlabel("월")
plt.ylabel("매출액 (원)")
plt.show()
```

```python
# '총매출' 열 추출
total_sales = df.loc['총매출']

# '목표매출' 열 추출
target_sales = df.loc['목표매출']

# 월별 목표매출 대비 달성률 계산
monthly_achievement_rate = (total_sales / target_sales) * 100

# 그래프 그리기
plt.plot(monthly_achievement_rate, label="월별 목표매출 대비 달성률")
plt.title("월별 목표매출 대비 달성률 그래프")
plt.xlabel("월")
plt.ylabel("달성률 (%)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 월별 광고비용과 소셜네트워크비용 추출
monthly_advertising_costs = df.loc["광고비용"].tolist()  # "월별 광고비용"은 실제 데이터에 맞게 선택
monthly_social_network_costs = df.loc["소셜네트워크비용"].tolist()  # "월별 소셜네트워크비용"은 실제 데이터에 맞게 선택

# 그래프 그리기
plt.plot(monthly_advertising_costs, label="월별 광고비용")
plt.plot(monthly_social_network_costs, label="월별 소셜네트워크비용")
plt.title("월별 광고비용과 소셜네트워크비용 그래프")
plt.xlabel("월")
plt.ylabel("비용 (원)")
plt.legend()
plt.show()
```

```python
# 월별 1온스별단가 추출
monthly_price_per_ounce = df.loc["1온스별단가"].to_list()

# 그래프 그리기
plt.plot(monthly_price_per_ounce, label="월별 1온스별단가")
plt.title("월별 1온스별단가 그래프")
plt.xlabel("월")
plt.ylabel("단가 (달러)")
plt.show()
```
